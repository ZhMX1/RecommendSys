## Model: user_product_rnn_mlp

* 将user和product（title和summary先过rnn层再与其余concat）分别过mlp层再concat再过mlp层

```python
import pandas as pd
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle

# 设置中文字体（以 SimHei 为例）
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False  # 显示负号

import numpy as np
import pandas as pd
from keras.models import Model
from keras.layers import Input, LSTM, Dense, Concatenate, Embedding, Flatten, RNN, Multiply, Dot,Reshape
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Dropout
from sklearn.metrics import mean_squared_error
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# 读取 Parquet 文件
df = pd.read_parquet('data_train.parquet')

long_summary = pd.read_parquet('long_summary.parquet')
df = pd.merge(df,long_summary,on="productId_encode",how="left")

# 显示 DataFrame
display(df)

column_names = df.columns.tolist()
print(column_names)
```

```python
# user variable
user_columns =  ['userId_encode']+ [elem for elem in column_names if elem.startswith('his_')]

# product variable (without text)
product_nontext_columns = ['productId_encode','train_mean_score','helpfulness_helpful','helpfulness_total']+[elem for elem in column_names if elem.startswith('categories_')]+[elem for elem in column_names if elem.startswith('brand_')]

# target
target_column = 'score'
```

```python
# title文本预处理
tokenizer_title = Tokenizer()
tokenizer_title.fit_on_texts(df.sample(20000)['title'].values.flatten())
sequences_title = tokenizer_title.texts_to_sequences(df['title'])

# 填充序列
max_length_title = 50  # 根据需要调整
X_product_title = pad_sequences(sequences_title, maxlen=max_length_title)

# 保存 Tokenizer
with open('tokenizer_title.pkl', 'wb') as f:
    pickle.dump(tokenizer_title, f)
```

```python
# summary文本预处理
tokenizer_summary = Tokenizer()
tokenizer_summary.fit_on_texts(df.sample(20000)['long_summary'].values.flatten())
sequences_summary = tokenizer_summary.texts_to_sequences(df['long_summary'])

# 填充序列
max_length_summary = 100  # 根据需要调整
X_product_summary = pad_sequences(sequences_summary, maxlen=max_length_summary)

# 保存 Tokenizer
with open('tokenizer_summary.pkl', 'wb') as f:
    pickle.dump(tokenizer_summary, f)
```

```python
X_user = df[user_columns].values
X_product_nontext = df[product_nontext_columns].values
```

```python
input_user = Input(shape=(X_user.shape[1],))
input_product_nontext = Input(shape=(X_product_nontext.shape[1],))
input_product_title = Input(shape=(X_product_title.shape[1],))
input_product_summary = Input(shape=(X_product_summary.shape[1],))

# user MLP 层
x_user = Dense(64, activation='relu')(input_user)
x_user = BatchNormalization()(x_user)  # 添加Batch Normalization
x_user = Dropout(0.5)(x_user)  # 在全连接层后添加Dropout
x_user = Dense(32, activation='relu')(x_user)
x_user = BatchNormalization()(x_user)  # 添加Batch Normalization
x_user = Dropout(0.5)(x_user)  # 在全连接层后添加Dropout

# product title RNN层
embedding_dim = 32
x_product_title = Embedding(input_dim=len(tokenizer_title.word_index) + 1, output_dim=embedding_dim)(input_product_title)
x_product_title = LSTM(16)(x_product_title)
x_product_title = BatchNormalization()(x_product_title) 
x_product_title = Dropout(0.5)(x_product_title)  # 添加Dropout层，设置50%的丢弃率

# product summary RNN层
embedding_dim = 16
x_product_summary = Embedding(input_dim=len(tokenizer_summary.word_index) + 1, output_dim=embedding_dim)(input_product_summary)
x_product_summary = LSTM(16)(x_product_summary)
x_product_summary = BatchNormalization()(x_product_summary) 
x_product_summary = Dropout(0.5)(x_product_summary)  # 添加Dropout层，设置50%的丢弃率

x_product = Concatenate()([x_product_title, x_product_summary, input_product_nontext])

# product MLP 层
x_product = Dense(64, activation='relu')(x_product)
x_product = BatchNormalization()(x_product)  # 添加Batch Normalization
x_product = Dropout(0.5)(x_product)  # 在全连接层后添加Dropout
x_product = Dense(32, activation='relu')(x_product)
x_product = BatchNormalization()(x_product)  # 添加Batch Normalization
x_product = Dropout(0.5)(x_product)  # 在全连接层后添加Dropout

# 拼接
x = Concatenate()([x_user, x_product])

# MLP 层
x = Dense(128, activation='relu')(x)
x = BatchNormalization()(x)  # 添加Batch Normalization
x = Dropout(0.5)(x)  # 在全连接层后添加Dropout
x = Dense(64, activation='relu')(x)
x = BatchNormalization()(x)  # 添加Batch Normalization
x = Dropout(0.5)(x)  # 在全连接层后添加Dropout
x = Dense(32, activation='relu')(x)
output = Dense(1)(x)  # 输出层

# 构建模型
model = Model(inputs=[input_user,input_product_nontext,input_product_title, input_product_summary], outputs=output)
model.summary()

# 编译模型
model.compile(optimizer='adam', loss='mean_squared_error')

# 训练模型
early_stopping = EarlyStopping(monitor='val_loss', patience=5)
history = model.fit([X_user,X_product_nontext,X_product_title,X_product_summary], df[target_column].values, epochs=50, batch_size=256, validation_split=0.2, callbacks=[early_stopping])
```

```python
# 绘制损失曲线
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
model.save('user_product_rnn_mlp.h5')
```
